```python
data_title = df_astro.title.values.tolist()
# remove distracting characters
data_title = [re.sub("\'", "", sent) for sent in data_title]
data_title = [re.sub("$", "", sent) for sent in data_title]
data_title = [re.sub("{", "", sent) for sent in data_title]
data_title = [re.sub("}", "", sent) for sent in data_title]
data_title = [re.sub("}", "", sent) for sent in data_title]
```

```python
# Remove new line characters
data_title = [re.sub('\s+', ' ', sent) for sent in data_title]
data_title = [re.sub('\n', ' ', sent) for sent in data_title]
```

```python
data_title
```

```python
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))  # deacc=True removes punctuations

data_title_words = list(sent_to_words(data_title))

print(data_title_words[:1])
```

## Create Bigram and Trigram models


```python
# Build the bigram and trigram models
bigram = gensim.models.Phrases(data_title_words, min_count=5, threshold=100) # higher threshold fewer phrases.
trigram = gensim.models.Phrases(bigram[data_title_words], threshold=100)  

# Faster way to get a sentence clubbed as a trigram/bigram
bigram_mod = gensim.models.phrases.Phraser(bigram)
trigram_mod = gensim.models.phrases.Phraser(trigram)

```

```python
# See trigram example
# print(trigram_mod[bigram_mod[data_title_words[0]]])
```

```python
# Define functions for stopwords, bigrams, trigrams and lemmatization
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) if word not in stop_words] for doc in texts]

def make_bigrams(texts):
    return [bigram_mod[doc] for doc in texts]

def make_trigrams(texts):
    return [trigram_mod[bigram_mod[doc]] for doc in texts]

def lemmatization(texts, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):
    """https://spacy.io/api/annotation"""
    texts_out = []
    for sent in texts:
        doc = nlp(" ".join(sent)) 
        texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
    return texts_out
```

```python
# Remove Stop Words
data_words_nostops = remove_stopwords(data_title_words)

# Form Bigrams
data_words_bigrams = make_bigrams(data_words_nostops)

```

```python
# Initialize spacy 'en' model, keeping only tagger component (for efficiency)
#!python3 -m spacy download en
nlp = spacy.load('en', disable=['tagger', 'textcat', 'parser', 'ner'])
```

```python
data_lemmatized = nlp(data_words_nostops)
```

```python
def lemmatize_corpus(nlp, corpus):
    return [[w.lemma_ for w in nlp(' '.join(doc))] for doc in corpus]
```

```python
# Do lemmatization keeping only noun, adj, vb, adv

data_lemmatized = lemmatize_corpus(nlp, data_words_nostops) #, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV'])

print(data_lemmatized[:1])
```

## Create the Dictionary and Corpus needed for Topic Modeling

```python
data_words_nostops
```

```python
# Create Dictionary
id2word = corpora.Dictionary(data_lemmatized)

# Create Corpus
texts = data_lemmatized

# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]

# View
print(corpus[:2])
```

```python
# Human readable format of corpus (term-frequency)
[[(id2word[id], freq) for id, freq in cp] for cp in corpus[:1]]
```

```python
# Build LDA model
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=20, 
                                           random_state=100,
                                           update_every=1,
                                           chunksize=100,
                                           passes= 2,
                                           alpha='auto',
                                           per_word_topics=True)
```

```python
# Print the Keyword in the 10 topics
print(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

```python
# Visualize the topics
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(lda_model, corpus, id2word)
vis
```

```python
# Build LDA model
lda_model_10 = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=10, 
                                           random_state=100,
                                           update_every=1,
                                           chunksize=100,
                                           passes= 2,
                                           alpha='auto',
                                           per_word_topics=True)

# Visualize the topics
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(lda_model_10, corpus, id2word)
vis
```
